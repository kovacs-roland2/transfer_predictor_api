```python
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

base_df = pd.DataFrame(columns=['Name','Age', 'ID', 'Team & Contract','Season'])

num = 1
for link in links:
  r = requests.get(link, headers=header)
  df = pd.read_html(r.text)[0]
  df = df[['Name','Age', 'ID', 'Team & Contract']]
  df['Season'] = season

  base_df = base_df.append(df)
  print(f'{num} of {len(links)} ready')
  num += 1

base_df.to_excel(f"/Users/{os.environ['USER']}/Documents/Uni/Szakdoga/Adatok/contract_expiry/{season}/.xlsx")
```
