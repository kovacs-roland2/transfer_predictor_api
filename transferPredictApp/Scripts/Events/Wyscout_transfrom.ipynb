```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
import pandasql as ps
import os
```

```python
df = pd.read_json(f"/Users/{os.environ['USER']}/Documents/Uni/Szakdoga/Adatok/wyscout/events/events_Spain.json")
df.head()
```

```python
player_df = pd.DataFrame(columns=['player_id', 'matches_played', 'attacking_duel', 'defending_duel', 'air_duel', 'loose_ball_duel', 'fouls ', 'gk_leaving_line',
    'offsides', 'acceleration', 'clearance', 'passes', 'crosses', 'smart_pass',
    'high_pass', 'shots'])

players = list(set(df.playerId))

num = 1

for player in players:
    current_player_df = df[df.playerId == player]

    player_id = player
    matches_played = len(set(current_player_df.matchId))

    #attacking duel, defending duel, air duel, loose ball duel
    attacking_duel = len(current_player_df[current_player_df.subEventName == 'Ground attacking duel'])
    defending_duel = len(current_player_df[current_player_df.subEventName == 'Ground defending duel'])
    air_duel = len(current_player_df[current_player_df.subEventName == 'Air duel'])
    loose_ball_duel = len(current_player_df[current_player_df.subEventName == 'Ground loose ball duel'])

    #fouls
    fouls = len(current_player_df[current_player_df.eventName == 'Foul'])

    #gk_leaving_line
    gk_leaving_line = len(current_player_df[current_player_df.eventName == 'Goalkeeper leaving line'])

    #offsides
    offsides = len(current_player_df[current_player_df.eventName == 'Offside'])

    #acceleration, clearance
    acceleration = len(current_player_df[current_player_df.subEventName == 'Acceleration'])
    clearance = len(current_player_df[current_player_df.subEventName == 'Clearance'])

    #passes, crosses, smart_pass, high_pass
    passes = len(current_player_df[current_player_df.eventName == 'Pass'])
    crosses = len(current_player_df[current_player_df.subEventName == 'Cross'])
    smart_pass = len(current_player_df[current_player_df.subEventName == 'Smart pass'])
    high_pass = len(current_player_df[current_player_df.subEventName == 'High pass'])

    #shots
    shots = len(current_player_df[current_player_df.eventName == 'Shot'])

    row_to_add = np.array([player_id, matches_played, attacking_duel, defending_duel, air_duel, loose_ball_duel, fouls , gk_leaving_line,
    offsides, acceleration, clearance, passes, crosses, smart_pass,
    high_pass, shots])

    data_to_append = {}
    for i in range(len(player_df.columns)):
        data_to_append[player_df.columns[i]] = row_to_add[i]
    player_df = player_df.append(data_to_append, ignore_index = True)

    print(f'{num} / {len(players)}')
    num += 1
```
