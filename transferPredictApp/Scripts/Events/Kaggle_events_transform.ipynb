```python
def event_name(row):
   if row['event_type'] == 0 :
      return 'Announcement'
   if row['event_type'] == 1 :
      return 'Attempt'
   if row['event_type'] == 2 :
      return 'Corner'
   if row['event_type'] == 3:
      return 'Foul'
   if row['event_type'] == 4:
      return 'Yellow card'
   if row['event_type'] == 5:
      return 'Second yellow card'
   if row['event_type'] == 6:
      return 'Red card'
   if row['event_type'] == 7:
      return 'Substitution'
   if row['event_type'] == 8:
      return 'Free kick won'
   if row['event_type'] == 9:
      return 'Offside'
   if row['event_type'] == 10:
      return 'Hand ball'
   if row['event_type'] == 11:
      return 'Penalty conceded'
```

```python
df['event_name'] = df.apply(lambda row: event_name(row), axis=1)
```

```python
def event_name2(row):
   if row['event_type2'] == 12 :
      return 'Key pass'
   if row['event_type2'] == 13 :
      return 'Failed through ball'
   if row['event_type2'] == 14 :
      return 'Sending off'
   if row['event_type2'] == 15:
      return 'Own goal'
```

```python
df['event_name2'] = df.apply(lambda row: event_name2(row), axis=1)
```

```python
def shot_place_name(row):
   if row['shot_place'] == 1 :
      return 'Bit too high'
   if row['shot_place'] == 2 :
      return 'Blocked'
   if row['shot_place'] == 3 :
      return 'Bottom left corner'
   if row['shot_place'] == 4:
      return 'Bottom right corner'
   if row['shot_place'] == 5:
      return 'Centre of the goal'
   if row['shot_place'] == 6:
      return 'High and wide'
   if row['shot_place'] == 7:
      return 'Hits the bar'
   if row['shot_place'] == 8:
      return 'Misses to the left'
   if row['shot_place'] == 9:
      return 'Misses to the right'
   if row['shot_place'] == 10:
      return 'Too high'
   if row['shot_place'] == 11:
      return 'Top centre of the goal'
   if row['shot_place'] == 12:
      return 'Top left corner'
   if row['shot_place'] == 13:
      return 'Top right corner'
```

```python
df['shot_place_name'] = df.apply(lambda row: shot_place_name(row), axis=1)
```

```python
def shot_outcome_name(row):
   if row['shot_outcome'] == 1 :
      return 'On target'
   if row['shot_outcome'] == 2 :
      return 'Off target'
   if row['shot_outcome'] == 3 :
      return 'Blocked'
   if row['shot_outcome'] == 4:
      return 'Hit the bar'
```

```python
df['shot_outcome_name'] = df.apply(lambda row: shot_outcome_name(row), axis=1)
```

```python
def location_name(row):
   if row['location'] == 1 :
      return 'Attacking half'
   if row['location'] == 2 :
      return 'Defensive half'
   if row['location'] == 3 :
      return 'Centre of the box'
   if row['location'] == 4:
      return 'Left wing'
   if row['location'] == 5:
      return 'Right wing'
   if row['location'] == 6:
      return 'Difficult angle and long range'
   if row['location'] == 7:
      return 'Difficult angle on the left'
   if row['location'] == 8:
      return 'Difficult angle on the right'
   if row['location'] == 9:
      return 'Left side of the box'
   if row['location'] == 10:
      return 'Left side of the six yard box'
   if row['location'] == 11:
      return 'Right side of the box'
   if row['location'] == 12:
      return 'Right side of the six yard box'
   if row['location'] == 13:
      return 'Very close range'
   if row['location'] == 14:
      return 'Penalty spot'
   if row['location'] == 15:
      return 'Outside the box'
   if row['location'] == 16:
      return 'Long range'
   if row['location'] == 17:
      return 'More than 35 yards'
   if row['location'] == 18:
      return 'More than 40 yards'
   if row['location'] == 19:
      return 'Not recorded'
```

```python
df['location'] = df.apply(lambda row: location_name(row), axis=1)
```

```python
def body_part_name(row):
   if row['bodypart'] == 1 :
      return 'Right foot'
   if row['bodypart'] == 2 :
      return 'Left foot'
   if row['bodypart'] == 3 :
      return 'Head'
```

```python
df['body_part_name'] = df.apply(lambda row: body_part_name(row), axis=1)
```

```python
def assist_method_name(row):
   if row['assist_method'] == 0 :
      return 'None'
   if row['assist_method'] == 1 :
      return 'Pass'
   if row['assist_method'] == 2 :
      return 'Cross'
   if row['assist_method'] == 3 :
      return 'Head'
   if row['assist_method'] == 4 :
      return 'Through ball'      
```

```python
df['assist_method_name'] = df.apply(lambda row: assist_method_name(row), axis=1)
```

```python
def situation_name(row):
   if row['situation'] == 1 :
      return 'Open play'
   if row['situation'] == 2 :
      return 'Set piece'
   if row['situation'] == 3 :
      return 'Corner'
   if row['situation'] == 4 :
      return 'Free kick'
```

```python
#join tables with train df
q1 = """SELECT DF.*,
        m.date,
        m.season,
        m.country,
        m.ht,
        m.at,
        m.fthg,
        m.ftag

        FROM df DF 

        LEFT JOIN df_season m
        ON DF.id_odsp	 = m.id_odsp	

        """

kaggle_df = ps.sqldf(q1, locals())
```

```python
kaggle_df.info()
```

```python
def side_name(row):
   if row['side'] == 1 :
      return row.ht
   if row['side'] == 2 :
      return row.at
```

```python
kaggle_df['club'] = kaggle_df.apply(lambda row: side_name(row), axis=1)
```

```python
kaggle_2012 = kaggle_df[kaggle_df.season == 2012]
kaggle_2013 = kaggle_df[kaggle_df.season == 2013]
kaggle_2014 = kaggle_df[kaggle_df.season == 2014]
kaggle_2015 = kaggle_df[kaggle_df.season == 2015]
kaggle_2016 = kaggle_df[kaggle_df.season == 2016]
kaggle_2017 = kaggle_df[kaggle_df.season == 2017]
```

```python
# player_df = pd.DataFrame(columns=['player', 'club', 'season', 'country','matches_played', 'shots', 'bar_hitted', 'shots_on_target'
#     , 'goals', 'shot_open_play', 'goals_open_play',
#     'inside_six_box', 'inside_box', 'outside_box', 'long_range', 'right_shot', 'left_shot'
#     , 'header', 'fouls', 'free_kick_won', 'free_kick_won_att_half',
#     'yellow_card', 'red_card'])

players = list(set(kaggle_2017.player))

num = 1

for player in players:
    if player is not None:
        current_player_df = kaggle_2017[kaggle_2017.player == player]

        player = player
        try:
            club = current_player_df.club.iloc[0]
        except:
            club = ''
        season = current_player_df.season.iloc[0]
        country = current_player_df.country.iloc[0]
        matches_played = len(set(current_player_df.id_odsp))

        #shots, bar_hitted, shots_on_goal, goals, inside_six_box, inside_box, outside_box, long_range
        #right_shot, left_shot, header, shot_open_play, goals_open_play
        shots = len(current_player_df[current_player_df.event_name == 'Attempt'])
        bar_hitted = len(current_player_df[current_player_df.shot_place_name == 'Hits the bar'])
        shots_on_target = len(current_player_df[current_player_df.shot_outcome_name == 'On target'])
        goals = len(current_player_df[current_player_df.is_goal == 1])
        shot_open_play = len(current_player_df[(current_player_df.event_name == 'Attempt') & (current_player_df.situation_name == 'Open play')])
        goals_open_play = len(current_player_df[(current_player_df.is_goal == 1) & (current_player_df.situation_name == 'Open play')])

        box_center = len(current_player_df[current_player_df.location == 'Centre of the box'])
        dif_long_range = len(current_player_df[current_player_df.location == 'Difficult angle and long range'])
        dif_left = len(current_player_df[current_player_df.location == 'Difficult angle on the left'])
        dif_right = len(current_player_df[current_player_df.location == 'Difficult angle on the right'])
        left_box = len(current_player_df[current_player_df.location == 'Left side of the box'])
        left_six_box = len(current_player_df[current_player_df.location == 'Left side of the six yard box'])
        long = len(current_player_df[current_player_df.location == 'Long range'])
        more_35 = len(current_player_df[current_player_df.location == 'More than 35 yards'])
        more_45 = len(current_player_df[current_player_df.location == 'More than 40 yards'])
        out_box = len(current_player_df[current_player_df.location == 'Outside the box'])
        pen = len(current_player_df[current_player_df.location == 'Penalty spot'])
        right_box = len(current_player_df[current_player_df.location == 'Right side of the box'])
        right_six_box = len(current_player_df[current_player_df.location == 'Right side of the six yard box'])
        six_box = len(current_player_df[current_player_df.location == 'Very close range'])

        inside_six_box = left_six_box + right_six_box + six_box
        inside_box = inside_six_box + box_center + left_box + right_box + pen
        outside_box = dif_long_range + long +more_35 + more_45 + out_box
        long_range = dif_long_range + long + more_35 + more_45

        right_shot = len(current_player_df[current_player_df.body_part_name == 'Left foot'])
        left_shot = len(current_player_df[current_player_df.body_part_name == 'Right foot'])
        header = len(current_player_df[current_player_df.body_part_name == 'Head'])

        #foul
        fouls = len(current_player_df[current_player_df.event_name == 'Foul'])

        #free kick won
        free_kick_won = len(current_player_df[current_player_df.event_name == 'Free kick won'])
        free_kick_won_att_half = len(current_player_df[(current_player_df.event_name == 'Free kick won') & (current_player_df.location == 'Attacking half')])

        #yellow card, red_card
        yellow_card = len(current_player_df[(current_player_df.event_name == 'Yellow card')])
        red_card = len(current_player_df[(current_player_df.event_name == 'Red card')])

        row_to_add = np.array([player, club, season, country, matches_played, shots, bar_hitted, shots_on_target, goals, shot_open_play, goals_open_play,
            inside_six_box, inside_box, outside_box, long_range, right_shot, left_shot, header, fouls, free_kick_won, free_kick_won_att_half,
            yellow_card, red_card])

        data_to_append = {}
        for i in range(len(player_df.columns)):
            data_to_append[player_df.columns[i]] = row_to_add[i]
        player_df = player_df.append(data_to_append, ignore_index = True)

        print(f'{num} / {len(players)}')
        num += 1
```
